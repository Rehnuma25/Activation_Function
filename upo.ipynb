```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
# Activation functions
def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha*x)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def swish(x):
    return x * sigmoid(x)

def gelu(x):
    return 0.5 * x * (1 + np.tanh(np.sqrt(2/np.pi) * (x + 0.044715 * x**3)))

```

```python
# List of functions
activations = {
    "ReLU": relu,
    "Leaky ReLU": leaky_relu,
    "Sigmoid": sigmoid,
    "Tanh": tanh,
    "Swish": swish,
    "GELU": gelu
}
```

```python
# User input
user_input = float(input("Enter a numeric value: "))

print("\nActivation Function Outputs:")
for name, func in activations.items():
    output = func(user_input)
    print(f"{name}: {output}")
```
